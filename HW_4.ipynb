### Уровень 0:


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.

Первая производная:

$f'(x) = (x^5)' + (4sin(2x))' + (cos(3x+3))' =$

$5x^4 + 4sin(2x)(2x)' - sin(3x+3)(3x+3)' =$

$5x^4 + 4(cos(2x)(2)) - sin(3x+3)(3) =$

$5x^4 + 8cos(2x) - 3sin(3x+3)$

Вторая производная:
$f''(x) = (5x^4 + 8cos(2x) - 3sin(3x+3)'=$

$20x^3 - 16sin(2x) - 9cos(3x+3)$

```python
import numpy as np

x = 1
der_1 = 5 * (x ** 4) + 8 * np.cos(2 * x) - 3 * np.sin(3 * x + 3)
der_2 = 20 * (x ** 3) - 16 * np.sin(2 * x) - 9 * np.cos(3 * x + 3)
print(f'При x = 1\nПервая производная: {der_1}\nВторая производная: {der_2}')
```

### Уровень 1:


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

Первая производная:

$f'(x) = (sin^5(2x+1))' = 5sin^4(2x+1)(sin(2x+1))'= $

$5sin^4(2x+1)cos(2x+1)(2x+1)'= $

$10sin^4(2x+1)cos(2x+1)$

Вторая производная:

$f''(x) = ((sin^5(2x+1)')' = (10sin^4(2x+1)cos(2x+1))' =$

$10(sin^4(2x+1))'cos(2x+1) + sin^4(2x+1)(cos(2x+1))')=$

$10(4sin^3(2x+1)(sin(2x+1))'cos(2x+1) - sin^4(2x+1)sin(2x+1)(2x+1)')=$

$10(8sin^3(2x+1)cos^2(2x+1)(2x+1)' - 2sin^5(2x+1)) =$

$80sin^3(2x+1)cos^2(2x+1) - 20sin^5(2x+1)$


На отрезке [-5; 5].

```python
def f(x):
    return np.sin(2 * x + 1) ** 5
def der_1(x):
    return 10 * np.sin(2 * x + 1) ** 5 * np.cos(2 * x + 1)
def der_2(x):
    return 80 * np.sin(2 * x + 1) ** 3 * np.cos(2 * x + 1) ** 2 - 20 * np.sin(2 * x + 1) ** 5
```

```python
x = np.arange(-5, 6, 0.1)
ders_1 = [der_1(d) for d in x]
ders_2 = [der_2(d) for d in x]
```

```python
import matplotlib.pylab as plt
x = x
plt.plot(np.sin(2 * x + 1) ** 5, label='$f(x) = sin^5(2x+1)$')
plt.plot(ders_1, label='1st derivative')
plt.plot(ders_2, label='2nd derivative')
plt.axis('tight')
plt.legend()
plt.show()


```
